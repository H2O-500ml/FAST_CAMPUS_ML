```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/bikeshare.csv')
```

```python
data.head(10)
```

```python
data.dtypes
```

```python
datetime = pd.DatetimeIndex(data['datetime'])
```

```python
data.dtypes
```

```python
data.head(10)
```

```python
data['year'] = datetime.year
```

```python
data['month'] = datetime.month
data['hour'] = datetime.hour
```

```python
data.drop('datetime', axis=1, inplace=True)
```

```python

data.head(10)
```

```python
data.rename(columns={'count':'total'}, inplace=True)
```

```python
data.head(1)
```

```python
data.drop(['casual','registered'], axis = 1, inplace = True)
```

```python

data.head(10)
```

```python
data['daytime'] = ((data['hour']>6)&(data['hour']<21)).astype(int)
```

```python
data.head(10)
```

```python
season_dummies = pd.get_dummies(data.season, prefix = 'season')
season_dummies
```

```python
season_dummies.drop(season_dummies.columns[0], axis=1, inplace=True)
data = pd.concat([data, season_dummies], axis=1)

data.drop('season', axis=1, inplace = True)
```

```python
data.head(10)
```

```python
hour_dummies=pd.get_dummies(data.hour, prefix='hour')
hour_dummies.drop(hour_dummies.columns[0], axis = 1, inplace = True)
data = pd.concat([data, hour_dummies], axis = 1)

data.drop('hour', axis = 1, inplace = True)
```

```python
data.head(10)
```

```python
print(data.columns)
```

```python

data.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('total', axis = 1)
Y = data.total
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=5079)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train, Y_train)
```

```python
Y_train_hat = linreg.predict(X_train)
Y_test_hat = linreg.predict(X_test)
```

```python
rmse_train = np.sqrt(mean_squared_error(Y_train, Y_train_hat))
rmse_test = np.sqrt(mean_squared_error(Y_test, Y_test_hat))
r2_train = linreg.score(X_train, Y_train)
r2_test = linreg.score(X_test, Y_test)

print('train rmse:', rmse_train)
print('test rmse:', rmse_test)
print('train r2:', r2_train)
print('test r2:', r2_test)
```

```python
ridge = Ridge(alpha=0.5, normalize=False)
ridge.fit(X_train, Y_train)

# Calculate Y_train_hat and Y_test_hat
Y_train_hat = ridge.predict(X_train)
Y_test_hat  = ridge.predict(X_test)

rmse_train = np.sqrt(mean_squared_error(Y_train, Y_train_hat))
rmse_test = np.sqrt(mean_squared_error(Y_test, Y_test_hat))
r2_train = ridge.score(X_train, Y_train)
r2_test = ridge.score(X_test, Y_test)
print("Train RMSE: ", rmse_train)
print("Test RMSE: ", rmse_test)
print("Train R^2: ", r2_train)
print("Test R^2: ", r2_test)
```

```python
lasso = Lasso(alpha = 1.0, normalize = False)
lasso.fit(X_train, Y_train)

Y_train_hat = lasso.predict(X_train)
Y_test_hat = lasso.predict(X_test)

rmse_train = np.sqrt(mean_squared_error(Y_train, Y_train_hat))
rmse_test = np.sqrt(mean_squared_error(Y_test, Y_test_hat))
r2_train = lasso.score(X_train, Y_train)
r2_test = lasso.score(X_test, Y_test)
print("Train RMSE: ", rmse_train)
print("Test RMSE: ", rmse_test)
print("Train R^2: ", r2_train)
print("Test R^2: ", r2_test)
```
